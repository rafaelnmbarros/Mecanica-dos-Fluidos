### EXEMPLO 7.6

As linhas de corrente para o escoamento em regime permanente de um fluido perfeito em torno do canto em 90º na Figura  7.20 são definidas pela função corrente c(x, y) = (5xy) m2 /s.  Determine a velocidade do escoamento no ponto x= 2 m, y= 3 m.  A equação de Bernoulli pode ser aplicada entre dois pontos  quaisquer dentro desse escoamento? 

Hibbeler, R. C.. Mecânica dos fluidos (p. 326). Edição do Kindle. 

```python
import numpy as np
import matplotlib.pyplot as plt

#Domínio de coordenadas
xmin=0
xmax=5
ymin=0
ymax=5

#Função para ciação de malha
#====================================================
def malha(xmin, xmax, ymin, ymax, nx, ny):
    x = np.linspace(xmin, xmax, nx)    #Coordenadas x dos pontos da malha
    y = np.linspace(ymin, ymax, ny)    #Coordenadas y dos pontos da malha
    X, Y = np.meshgrid(x, y)        #Criação da malha
    return X, Y

#Plotagem do mapa de cores da função de corrente
#====================================================
X, Y = malha(xmin, xmax, ymin, ymax, 50, 50)
corrente=5*X*Y                     #Função de corrente
fig = plt.figure(figsize=(8, 6))   #Tamanho da figura
ax = fig.add_subplot(111)          #Apenas uma figura
im = ax.contourf(X, Y,             #Plotagem da função de corrente
    corrente, cmap='plasma', levels=np.linspace(corrente.min(), corrente.max(), 9)) 
cbar=fig.colorbar(im)              #Barra de cores
cbar.set_label(                    #Rótulo da barra de cores
    label=r'$\frac{m^{3}/s}{m}$', size=20)

#Plotagem de vetores velocidade
#====================================================
X, Y = malha(xmin, xmax, ymin, ymax, 15, 15)
U=5*X                             #Componente x da velocidade
V=-5*Y                            #Componente y da velocidade
VR=(U**2+V**2)**0.5
velocidade=ax.quiver(X, Y, U, V, color='k', #Plotagem do vetor velocidade
    angles='xy', scale=250, width=0.004, pivot='mid')
qk=ax.quiverkey(velocidade, 0.7, 1.015, 10, label='10 m/s', labelpos='E', coordinates='figure')

#Edição do gráfico
#====================================================
plt.title('Escoamento em torno \n de um canto em 90°', fontsize=15)
plt.xlabel('x (m)', fontsize=15)
plt.ylabel('y (m)', fontsize=15)
plt.grid()
```
