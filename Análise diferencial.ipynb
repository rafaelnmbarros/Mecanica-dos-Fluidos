```python
import sympy as sp
from sympy import *
sp.init_printing(use_latex="mathjax")
from sympy.plotting import plot
from sympy import latex


#Entrada de dados
x, y, t, C, f_x, d_f_x = sp.symbols("x y t C f(x) f'(x)")
u=2*x**2
v=-4*x*y+x**2
valor_x=3
valor_y=1
xmin=0
xmax=8
ymin=0
ymax=8

psi=sp.integrate(u,y)+f_x               #integra√ß√£o em rela√ß√£o a y
d_f_x=-(v+sp.diff(psi,x))               #derivada da fun√ß√£o f(x)
definicao_f_x = sp.integrate(d_f_x,x)+C #integra√ß√£o em rela√ß√£o a x
psi=psi.subs(f_x,definicao_f_x)         #substitui√ß√£o da fun√ß√£o f(x)
psi=psi.subs(C,0)                       #utilizando C = 0
valor_psi=psi.subs(x,valor_x).subs(y,valor_y)
Eq_y=sp.solveset(sp.Eq(psi,valor_psi),y).args[0]

#Exibi√ß√£o dos resultados
p1=plot(Eq_y, label='Linha de trajet√≥ria', xlabel='x(m)', ylabel = 'y (m)', show=False, legend=True, line_color='blue', xlim=(xmin, xmax), ylim=(ymin, ymax))
p1.show()
```
