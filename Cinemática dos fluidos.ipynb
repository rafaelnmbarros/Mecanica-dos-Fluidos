### Exemplo 3.1 do livro Mecânica dos Fluidos (R. C. Hibbeler)

```python
import sympy as sp
from sympy import *
sp.init_printing(use_latex="mathjax")
from sympy.plotting import plot
from sympy import latex

#Entrada de dados
x, y, C=sp.symbols('x y C')
u=6*y
v=3
valor_x=1
valor_y=2

#Linha de trajetória
Eq_esq=sp.integrate(u,y)
Eq_dir=sp.integrate(v,x)
Eq1=Eq_esq-Eq_dir-C
valor_C=sp.solveset(Eq1.subs([(x,valor_x), (y,valor_y)]),C).args[0]
Eq2=Eq(Eq1.subs(C,valor_C))

#Exibição dos resultados
print("Equação da Linha de trajetória:")
display(Eq2)
p1=plot(sp.solveset(Eq2,y).args[0], label='Linha de corrente', xlabel='x(m)', ylabel = 'y (m)', show=False, legend=True, line_color='blue')
p1.show()
```

### Exemplo 3.2 do livro Mecânica dos Fluidos (R. C. Hibbeler)

```python
import sympy as sp
from sympy import *
sp.init_printing(use_latex="mathjax")
from sympy.plotting import plot

#Entrada de dados
x, y, t, C = sp.symbols('x y t C')
u=3+0*t
v=6*t
valor_t=2

#Linha de trajetória
Eq_x_t=sp.integrate(u,(t,0,t))
Eq_y_t=sp.integrate(v,(t,0,t))
valor_x=Eq_x_t.subs(t,valor_t)
valor_y=Eq_y_t.subs(t,valor_t)
Eq_t_x=sp.solveset(Eq(Eq_x_t,x),t).args[0]
Eq1=Eq_y_t.subs(t,Eq_t_x)

#Linha de corrente
Eq2=sp.integrate(v.subs(t,valor_t)/u.subs(t,valor_t),x)+C
valor_C=sp.solveset(Eq(Eq2,y).subs([(x,valor_x),(y,valor_y)]),C).args[0]
Eq2=Eq2.subs(C,valor_C)

#Exibição dos resultados
p1=plot(Eq1, label='Linha de trajetória', xlabel='x(m)', ylabel = 'y (m)', show=False, legend=True, line_color='blue')
p2=plot(Eq2, label='Linha de corrente para o instante especificado', xlabel='x(m)', ylabel = 'y (m)', show=False, line_color='red')
p1.append(p2[0])
p1.show()
```

 ### Problema 3-40 do livro Mecânica dos Fluidos (R. C. Hibbeler)

```python
import sympy as sp
from sympy import *
sp.init_printing(use_latex="mathjax")
from sympy.plotting import plot

#Entrada de dados
x, y, t, C = sp.symbols('x y t C')
u=4*y
v=2*x
valor_x=2
valor_y=1

V=(u**2+v**2)**0.5
ax=sp.diff(u,t)+u*sp.diff(u,x)+v*sp.diff(u,y)
ay=sp.diff(v,t)+u*sp.diff(v,x)+v*sp.diff(v,y)
valor_V=V.subs(x,valor_x).subs(y,valor_y)
valor_ax=ax.subs([(x,valor_x),(y,valor_y)])
valor_ay=ay.subs([(x,valor_x),(y,valor_y)])
valor_a=(valor_ax**2+valor_ay**2)**0.5

Eq1=sp.integrate(u,(y,valor_y,y))
Eq2=sp.integrate(v,(x,valor_x,x))
Eq3=Eq(Eq1,Eq2)
Eq4=sp.solveset(Eq3,y).args[0]
Eq5=sp.solveset(Eq3,y).args[1]

print('Velocidade (m/s):')
display(round(float(valor_V),2))
print('\n')
print('Aceleração (m/s2):')
display(round(float(valor_a),2))
print('\n')
print('Equação da linha de corrente:')
display(sp.solveset(Eq3,y))

p1=plot(Eq4, label='Linha de corrente', xlabel='x(m)', ylabel = 'y (m)', show=False, legend=False, line_color='blue', title='Linhas de Corrente', xlim=(0,10))
p2=plot(Eq5, label='Linha de corrente', xlabel='x(m)', ylabel = 'y (m)', show=False, legend=False, line_color='blue', title='Linhas de Corrente', xlim=(0,10))
p1.append(p2[0])
p1.show()
```

## Problema 3.24 (Brunetti)

```python
import numpy as np
import matplotlib.pyplot as plt

#Criação da malha para plotagem da função de corrente
X,Y = np.meshgrid(np.linspace(0,5,250), np.linspace(0,5,250))

#Dados de entrada
t=5
U=3+2*X*Y+4*t**2
V=X*Y**2+3*t
VR=(U**2+V**2)**0.5

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)

#Plotagem da função de corrente
im = ax.contourf(X, Y,             #Plotagem da função de corrente
    VR, cmap='cool', levels=np.linspace(VR.min(), VR.max(), 24))
ax.contour(im, colors='gray')
plt.clabel(im, inline=1, fontsize=12, colors='k')

cbar=fig.colorbar(im)              #Barra de cores

X,Y = np.meshgrid(np.linspace(0,5,10), np.linspace(0,5,10))
U=3+2*X*Y+4*t**2
V=X*Y**2+3*t
VR=(U**2+V**2)**0.5
velocidade=ax.quiver(X, Y,       #Plotagem do vetor velocidade
    U, V, angles='xy', scale=2000, width=0.0045, pivot='mid')
qk=ax.quiverkey(velocidade, 0.7, 0.9, 50, label='50 m/s', labelpos='E', coordinates='figure', )

plt.grid(linestyle='--', color='grey')
```
