```python
import matplotlib.pyplot as plt
import numpy as np


#Dados de entrada
t=5

#Área do gráfico
fig=plt.figure(figsize=(10,10))
ax=fig.add_subplot(111)

#Linhas de corrente
X,Y=np.meshgrid(np.linspace(0,5,250),np.linspace(0,5,250)) #Criação da malha
U=3+2*X*Y+4*t**2
V=X*Y**2+3*t
VR=(U**2+V**2)**0.5
LinhasDeCorrente=plt.contourf(X,Y,VR,levels=np.linspace(VR.min(),VR.max(),20),cmap='cool')
ax.contour(LinhasDeCorrente, colors='k')
plt.clabel(LinhasDeCorrente, colors='k', fontsize=12)
BarraDeCores=fig.colorbar(LinhasDeCorrente)


#Vetores de velocidade
X,Y=np.meshgrid(np.linspace(0,5,10),np.linspace(0,5,10)) #Criação da malha
U=3+2*X*Y+4*t**2
V=X*Y**2+3*t
Velocidades=plt.quiver(X,Y,U,V)
```
